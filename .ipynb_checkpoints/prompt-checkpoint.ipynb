### Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece.

**Investigating the Bar Coupons**

Now, we will lead you through an exploration of just the bar related coupons.  

1. Create a new `DataFrame` that contains just the bar coupons.


```python
df_bar = data[data['coupon'] == 'Bar']
df_bar
```

2. What proportion of bar coupons were accepted?


```python
len(df_bar[df_bar['Y'] == 1]) / len(df_bar)
```

3. Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.


```python
df_bar2 = df_bar[["Bar","Y"]]
df_bar2.loc[(df_bar2.Bar == "never") | (df_bar2.Bar == "1~3") | (df_bar2.Bar == "less1"), "Bar"] = "<= 3 Times"
df_bar2.loc[(df_bar2.Bar == "4~8") | (df_bar2.Bar == "gt8"), "Bar"] = "> 3 Times"
df_bar2 = df_bar2.groupby("Bar").agg({"Bar": 'size', 'Y':'sum'})
df_bar2.rename(columns={"Bar":"total_count","Y":"acceptance_count"},inplace=True)
df_bar2 = df_bar2.reset_index()
df_bar2["acceptance_percentage"] = df_bar2["acceptance_count"]/df_bar2["total_count"]*100
df_bar2
```

```python
sns.barplot(data=df_bar2, x='Bar', y='acceptance_percentage')
plt.title('Conversion Rate for Bar Goers')
plt.xlabel('Number of Times Visited a Bar')
plt.ylabel('Conversion Rate')
```

4. Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
df_bar3 = df_bar[['age', 'Bar', 'Y']].copy()
df_bar3['age'] = pd.to_numeric(df_bar3['age'])

df_bar3['25_older_visit_bar_once_or_more'] = np.where(
    (df_bar3['age'] > 25) & 
    ((df_bar3['Bar'] == '1~3') | (df_bar3['Bar'] == '4~8') | (df_bar3['Bar'] == 'gt8'))
    , "Yes", "No")

df_bar3 = df_bar3.groupby("25_older_visit_bar_once_or_more").agg({"25_older_visit_bar_once_or_more": 'size', 'Y':'sum'})
df_bar3.rename(columns={"25_older_visit_bar_once_or_more":"total_count","Y":"acceptance_count"},inplace=True)
df_bar3 = df_bar3.reset_index()
df_bar3["acceptance_percentage"] = df_bar3["acceptance_count"]/df_bar3["total_count"]*100
df_bar3
```

```python
sns.barplot(data=df_bar3, x='25_older_visit_bar_once_or_more', y='acceptance_percentage')
# Rotate the x-axis labels
plt.title('Conversion Rate for Bar Goers Over 25 Years Old')
plt.ylabel('Conversion Rate')
```

5. Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry. 


```python
df_bar4 = df_bar[['passanger', 'occupation', 'Bar', 'Y']].copy()

df_bar4['passanger_not_kid_occupation_not_FFF'] = np.where(
    (df_bar4['passanger'] != 'Kid(s)') & (df_bar4['occupation'] != 'Farming Fishing & Forestry') &
    ((df_bar4['Bar'] == '1~3') | (df_bar4['Bar'] == '4~8') | (df_bar4['Bar'] == 'gt8'))
    , "Yes", "No")

df_bar4 = df_bar4.groupby("passanger_not_kid_occupation_not_FFF").agg({"passanger_not_kid_occupation_not_FFF": 'size', 'Y':'sum'})
df_bar4.rename(columns={"passanger_not_kid_occupation_not_FFF":"total_count","Y":"acceptance_count"},inplace=True)
df_bar4 = df_bar4.reset_index()
df_bar4["acceptance_percentage"] = df_bar4["acceptance_count"]/df_bar4["total_count"]*100
df_bar4
```

```python
sns.barplot(data=df_bar4, x='passanger_not_kid_occupation_not_FFF', y='acceptance_percentage')
plt.title('Conversion Rate for Bar Goers no Kid(s) passanger not Farmer, Fisheries or Forestry')
plt.ylabel('Conversion Rate')
```

6. Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K. 



```python
# Extract the dataframe
df_bar5 = df_bar[["maritalStatus","age","income","passanger","Bar","occupation","RestaurantLessThan20","Y"]].copy()

# Set age column to numeric
df_bar5['age'] = pd.to_numeric(df_bar5['age'])

# Filter
df_bar5['recipient'] = np.where(
    (
        ((df_bar5['Bar'] == '1~3') | (df_bar5['Bar'] == '4~8') | (df_bar5['Bar'] == 'gt8')) & 
        (
            ((df_bar5['passanger'] != 'Kid(s)') & 
             (df_bar5['maritalStatus'] != 'Widowed')) |
            (df_bar5['age'] < 30) 
        )
    ) | 
    (
        ((df_bar5["income"] == 'Less than $12500') | (df_bar5["income"] == '$12500 - $24999') | (df_bar5["income"] == '$25000 - $37499') | (df_bar5["income"] == '$37500 - $49999')) &
        ((df_bar5['RestaurantLessThan20'] == '4~8') | (df_bar5['RestaurantLessThan20'] == 'gt8')) 
    )
    , "Yes", "No")

df_bar5 = df_bar5.groupby("recipient").agg({"recipient": 'size', 'Y':'sum'})
df_bar5.rename(columns={"recipient":"total_count","Y":"acceptance_count"},inplace=True)
df_bar5 = df_bar5.reset_index()
df_bar5["acceptance_percentage"] = df_bar5["acceptance_count"]/df_bar5["total_count"]*100
df_bar5
```

```python
sns.barplot(data=df_bar5, x='recipient', y='acceptance_percentage')
# Rotate the x-axis labels
plt.title('Conversion Rate')
plt.ylabel('Conversion Rate')
plt.xlabel('Accepted Bar coupon')
```

7.  Based on these observations, what do you hypothesize about drivers who accepted the bar coupons?

There is a strong indication that someone who frequents bars more that four times per month will accept the bar coupon, regardless of other factors investigated.  Also if you frequent cheap restaraunts more that four times per month and earn less than 50k there is a strong indication of accepting the coupon.

### Independent Investigation

Using the bar coupon example as motivation, you are to explore one of the other coupon groups and try to determine the characteristics of passengers who accept the coupons.  

We will explore the likelihood of expensive restaurant coupons being accepted

```python
df_rest = data[data['coupon'] == 'Restaurant(20-50)']
df_rest
```

What is the overall likelihood a restaurant costing betweeen 20-50 will be accepted?

```python
len(df_rest[df_rest['Y'] == 1]) / len(df_rest)
```

What is the likelihood someone driving alone with no urgent place to be is to accept the expensive restaraunt coupon?

```python
df_rest2 = df_rest[["destination","passanger","Y"]].copy()

df_rest2['recipient'] = np.where(
    (df_rest2['passanger'] == 'Alone') & 
    (df_rest2['destination'] == 'No Urgent Place')
    , "Yes", "No")

df_rest2 = df_rest2.groupby("recipient").agg({"recipient": 'size', 'Y':'sum'})
df_rest2.rename(columns={"recipient":"total_count","Y":"acceptance_count"},inplace=True)
df_rest2 = df_rest2.reset_index()
df_rest2["acceptance_percentage"] = df_rest2["acceptance_count"]/df_rest2["total_count"]*100
df_rest2
```

```python
sns.barplot(data=df_rest2, x='recipient', y='acceptance_percentage')
# Rotate the x-axis labels
plt.title('Conversion Rate')
plt.ylabel('Conversion Rate')
plt.xlabel('Accepted Restaurant coupon')
```

What is the likelihood someone with no urgent place to be who goes to expensive resaraunts more then three time per month is to accept the expensive restaraunt coupon?

```python
df_rest3 = df_rest[["destination","passanger","Restaurant20To50","Y"]].copy()

df_rest3['recipient'] = np.where(
    (df_rest3['destination'] == 'No Urgent Place') &
    ((df_rest3['Restaurant20To50'] == '4~8') | (df_rest3['Restaurant20To50'] == 'gt8'))
    , "Yes", "No")

df_rest3 = df_rest3.groupby("recipient").agg({"recipient": 'size', 'Y':'sum'})
df_rest3.rename(columns={"recipient":"total_count","Y":"acceptance_count"},inplace=True)
df_rest3 = df_rest3.reset_index()
df_rest3["acceptance_percentage"] = df_rest3["acceptance_count"]/df_rest3["total_count"]*100
df_rest3
```

```python
sns.barplot(data=df_rest3, x='recipient', y='acceptance_percentage')
# Rotate the x-axis labels
plt.title('Conversion Rate')
plt.ylabel('Conversion Rate')
plt.xlabel('Accepted Restaurant coupon')
```

Here we find someone driving with no urgent place to be who frequents expensive restaurants more than three times per month has a signficantly higher acceptance rate than those who don't frequent expensive restaurants
